## Permutation test

```python
import json 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from hypyp import stats 
'''
Comparing inter-connectivity to random signal
'''

f = open('results_plv.json')

# returns JSON object as 
# a dictionary
results = json.load(f)
  
# Iterating through the json
# list
fp1 = []
sf1 = []
fp2 = []
sf2 = []
reunion = []

all_stages = [fp1, sf1, fp2, sf2, reunion]
for part, data in results.items():
    for stage, bands in data.items():
        stage = stage[1]
        all_stages[int(stage)-1].append(bands)

fp1_theta = []
fp1_alpha = []
sf1_theta = []
sf1_alpha = []
fp2_theta = []
fp2_alpha = []
sf2_theta = []
sf2_alpha = []
ru_theta = []
ru_alpha = []
for part1, part3, part4, part5 in zip(fp1, fp2, sf2, reunion):
    fp1_theta.append(np.mean(part1['theta']))
    fp1_alpha.append(np.mean(part1['alpha']))

for part2 in sf1:
    sf1_theta.append(np.mean(part2['theta']))
    sf1_alpha.append(np.mean(part2['alpha']))

for part3 in fp2:
    fp2_theta.append(np.mean(part3['theta']))
    fp2_alpha.append(np.mean(part3['alpha']))

for part4 in sf2:
    sf2_theta.append(np.mean(part4['theta']))
    sf2_alpha.append(np.mean(part4['alpha']))

for part5 in reunion:
    ru_theta.append(np.mean(part5['theta']))
    ru_alpha.append(np.mean(part5['alpha']))


data_theta = [np.vstack(fp1_theta), np.vstack(sf1_theta), np.vstack(fp2_theta), np.vstack(sf2_theta), np.vstack(ru_theta)] # , np.vstack(fp2_theta), np.vstack(sf2_theta), np.vstack(ru_theta)
data_alpha = [np.vstack(fp1_alpha), np.vstack(sf1_alpha), np.vstack(fp2_alpha), np.vstack(sf2_alpha), np.vstack(ru_alpha)]

# Closing file
f.close()
```

```python
permutation_test = stats.statscondCluster(data = data_theta,
                                          freqs_mean = np.arange(3, 7),
                                          ch_con_freq = None,
                                          tail = 0,
                                          n_permutations = 5000,
                                          alpha = 0.05)
```
