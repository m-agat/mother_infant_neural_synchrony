# Cross-frequency PLV - theta band

```python
import json 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from hypyp import stats 

f = open('results_theta.json')

# returns JSON object as 
# a dictionary
results = json.load(f)
  
# Iterating through the json
# list
fp1 = []
sf1 = []
fp2 = []
sf2 = []

all_stages = [fp1, sf1, fp2, sf2]
for part, stages in results.items():
    for stage, data in stages.items():
        stage = int(stage[1]) - 1
        if stage == 4:
            continue
        all_stages[stage].append(np.nanmean(data))

sf2 = [x for x in sf2 if str(x) != 'nan']

data_theta_plot = [fp1, sf1, fp2, sf2] 

fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
# ax.set_ylim(0.1, 0.22)
ax.set_title('Theta Band')
ax.set_ylabel('Values')
 # x-axis labels
ax.set_xticklabels(['free play 1', 'still face 1', 'free play 2', 'still face 2', 'reunion'])
 
# Creating plot
bp = ax.boxplot(data_theta_plot)

# show plot
plt.show()

f.close()
```

# wPLI

```python
import json 
import numpy as np
import json 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from hypyp import stats 

f = open('results_pli.json')

# returns JSON object as 
# a dictionary
results = json.load(f)

# Iterating through the json
# list
fp1 = []
sf1 = []
fp2 = []
sf2 = []
reunion = []

all_stages = [fp1, sf1, fp2, sf2, reunion]
for part, data in results.items():
    for stage, bands in data.items():
        stage = stage[1]
        all_stages[int(stage)-1].append(bands)

fp1_theta = []
fp1_alpha = []
sf1_theta = []
sf1_alpha = []
fp2_theta = []
fp2_alpha = []
sf2_theta = []
sf2_alpha = []
ru_theta = []
ru_alpha = []
for part1 in fp1:
    fp1_theta.append(np.mean(part1['theta']))
    fp1_alpha.append(np.mean(part1['alpha']))

for part2 in sf1:
    sf1_theta.append(np.mean(part2['theta']))
    sf1_alpha.append(np.mean(part2['alpha']))

for part3 in fp2:
    fp2_theta.append(np.mean(part3['theta']))
    fp2_alpha.append(np.mean(part3['alpha']))

for part4 in sf2:
    sf2_theta.append(np.mean(part4['theta']))
    sf2_alpha.append(np.mean(part4['alpha']))

for part5 in reunion:
    ru_theta.append(np.mean(part5['theta']))
    ru_alpha.append(np.mean(part5['alpha']))

data_theta_pli = [fp1_theta, sf1_theta, fp2_theta, sf2_theta, ru_theta]
data_alpha_pli = [fp1_alpha, sf1_alpha, fp2_alpha, sf2_alpha, ru_alpha]

fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
# ax.set_ylim(0.14, 0.23)
ax.set_title('Theta Band')
ax.set_ylabel('Values')
 # x-axis labels
ax.set_xticklabels(['free play 1', 'still face 1', 'free play 2', 'still face 2', 'reunion'])
 
# Creating plot
bp = ax.boxplot(data_theta_pli)


# show plot
plt.show()

# Closing file
f.close()
```
